# Logica de proposiciones
## Tablas de verdad

### objetivo de clase
* **comprender el concepto de tablas de verdad**
* **construir ejemplos de tablas de verdad**

### Definicion

Es una tabla que muestra el **valor de verdad** de una proposciion compuesta, asi como de algunos casos de proposiciones simples, cuando estas utilizan los **operadores logicos**

### Tabla de verdad de la negacion

| P | ~P |
|---|----|
| V | F  |
| F | V  |


| P | ~ (~P) |
|---|-------|
| V | V     |
| F | F     |

### Tabla de verdad conjuncion

### construccion de una tabla de verdad

* Las tablas de verdad toman relevancia al permitirnos exponer las relaciones de proposiciones logicas para su construccion seguimos los siguientes pasos:
    1. asignar variables a las proposiciones simples
    2. Realizar la traduccion logica de la propuesta
    3. obtener las combinaciones posibles 2^n
    4. Asignar los valores de verdad
    5. Resolver la operacion logica propuesta
