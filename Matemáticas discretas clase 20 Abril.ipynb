## Logica de Proposiciones

### Entender el concepto de proposiciones y variables
Una propocision o enunciado constituye una oracion que tiene un valor de verdad, es decir , puede ser verdadera o falsa pero no ambas.

## Eventos logicos & Compuertas logicas

### Que no es una proposicion
Si la oracion es una pregunta, una orden o carece de sentido o es muy imprecisa entonces no puede ser clasificada como verdadera o falsa y por lo tanto no puede ser una proposicion

## variables

### Variables numericas

1. En las matematicas utilizamos letras a,b,c,...,x,y,z para representar lo que conocemos como variables
2. dichas variables pueden ser reemplazadas con valores numericos
3. Asi tambien podemos combinar las variables con diversos operadores elementales como son:
    1. Suma
    2. Resta
    3. Multiplicacion
    4. Division


### Variables proposicionales

1. En el area de la logica las letras a,b,c se pueden utilizar como variables que puede ser reemplazadas por proposiciones simples
2. asi es posible utilizar una proposicion haciendo referencia solo a la variable proposicional


## tipos de proposiciones 

### Proposiciones simples

Son aquellas proposiciones que estan estructuradas por una unica oracion 

Ejemplos: 

    p: El oro es un metal precioso.
    q: Hoy es martes.
    r: Benito Juarez nacion en Oaxaca.
    s: Rodolfo Neri Vela fue el primer astronauta mexicano

### Propocisiones compuestas

Las propocisiones compuestas son aquellas que estan estructuradas por dos o mas proposiciones simples unidas por operadores logicos

Ejemplo:

    Pitagoras era griego y geometra
    Lo podemos descomponer en proposiciones simples
    Pitagoras era griego
    Pitagoras era geometra


Ejemplos:


   * El sentido de la calle es hacia el norte **(o)** hacia el sur.
   *   **(Si)** salgo tarde, **(entonces)** no visitare a la abuela
   * Ire al cine **(si y solo si)** tu pagas las palomitas
