# Matematicas discretas 
## logica de proposiciones 

### tautologia 

* **Repeticion innecesaria** y poco afotunada

* Enuncaido que, con otras palabras, **repite lo mismo** que ya se ha dicho, sin que aporte nueva informacio

* En la logica la tautologia es una **proposicion que es verdadera** para **todos los posibles valores de verdad** de sus componenetes simples

* **a <=> a siempre es verdadero**

* Basicamente son enunciados obvios y redundantes, en otras palabaras, explican y afirman a si mismos


* Su termino griego lo hace evidente al ser conformado por **tauto ( lo mismo) y logos (palabra o saber)**

* Algunos ejemplos de tautologia:

* Vive la vida

* Entrar adentro 

* Antecendentes previso

* proyecto de futuro

* el triangulo tiene tres lados

* un lleno completo

* sopresa inesperada

* corri la distancia con mis propios pies

* todo lo que esta de mas sobra

* las cosas cayeron hacia abajo

* Subi hacia arriba de la escalera

* El frio es causado por el deceson de la temperatura

* ejemplo de tabla de verdad de tautologia

| p | p <=> p |
|---|---------|
| V |    V    |
| F |    V    |

**Cualquier tabla en la que siempre encontramos verdadero en la salida**


### Contradiccion

* Conjunto de proposiciones que al oponerse reciprocamente se invalidan

* oposicion (contrariedad o antagonismo)

* al **contrario de las tautologias** , que son **ciertas en cualquier formulacion posible** 

* Las **contradicciones son falsas** sean cuales sean los valores de sus premisas, ya que en su estructura argumental se niega la conclusion que se desea obtener.


* ejemplos

* La vida es larga y es corta

* La amo y la odio al mismo tiempo

* Te estoy viendo y no te veo

* Nunca digas nunca

* la tarea es facil y muy dificil

* Caimos hacia las alturas

* **a <=> ~a**

* ejemplo de tabla de verdad de la contradiccion

| p | p <=> ~p  |
|---|-----------|
| V |     F     |
| F |     F     |
