## operadores logicos

Los operadores logicos son aquellos simbolos que permiten decidir que **valor de verdad tiene una proposicion**

**el valor de vedad de una proposicion compuesta es verdadero o falso** y depende de los valores de verdad de las **procposiciones simples** que la estructuran, las cuales estan combinadas por **operadores logicos**

### Negacion

* la negacion es representada con el simbolo (~)
* La negacion de una proposicion invierte el significado resultante
* la negacion de cualquier proposicion p sera falsa cuando se niegue una proposicion verdadera y verdadera cuando se niegue una falsa 


#### ejemplos

**No,nunca,ni,jamas, es falso, no es cierto, no ocurre, de ninguna forma, por nada de, en lo absoluto, etc**

* p: El acusado dice la verdad
* ~p: El acusado no dice la verdad
* En este caso, ~p tambien se puede traducir como "No es cierto que el acusado dice la verdad" o " es falso que el acusado dice la verdad"

### conjuncion

* el simbolo de la conjuncion es ^ y es la conguacion de dos valores resulta en verdadero solo si ambas proposiciones son verdaderas, y en falso de cualquier otra forma.

* entre sus propiedades debemos de recordar que la conjugacion es conmutativa, es decir: 

                                            (P ^ q) = (q ^ P)

### ejemplos

* p: el acusado es pobre
* q: el acusado es honesto

* p^q: el acusado es pobre y honesto

* r: el helio es mas liviano que el aire.
* s: el helio es explosivo
* r^s: el helio es mas liviano que el aire, el helio es explosivo

### disyuncion inclusiva

* el simbolo para la disyncion inclusiva es (V)
* si p y q representa dos proposiciones simples, entonces la proposicion compuesta:
                                                   p V q 
                                                   
* la disyuncion es conmutativa:
                                               p V q = q V p
                                               

#### ejemplos:

* r: llovera en la tarde
* s: saldra el sol
* r V s: Llovera en la tarde o saldra el sol


### disyuncion exclusiva

*  el simbolo para la disyuncion exclusiva es XOR

* si p y q representa dos proposiciones simples, entonces la proposicion compuesta p(XOR)q solo sera falsa cuando las dos proposiciones tuvieren el mismo valor

* la disyuncion exclusiva es conmutativa, es decir:
                                            p XOR q = q XOR p
                                            
                                            
#### ejemplo:

* t: apruebas el ciclo escolar
* u: repruebas el ciclo escolar
* tXORu: apruebas o repurebas el ciclo escolar

### condicional o implicacion (Se sigue)

* el simbolo para la condicional o implicacion es (=>)
* si p y q representan dos proposiciones simples entonces la proposicion compuesta:
                                        p => q
* solo sera falsa cuando p, llamado antecedente o hipotesis sea verdadero y q llamado consecuente o conclusion sea falso.

* Algunas formas de la condicional o implicacion son:
* si ... entonces, se sigue, portanto, se infiere, de ahi que, se deduce, implica, entre otras

* la condicional no es conmutativa es decir:
                                         p => q != q => p
                                         p => q => ~q => ~p

#### ejemplos

* t: llueve
* u: me mojare
* t => u: si llueve, entonces me mojare.

* p: estudio
* q: aprobare el ciclo escolar
* p=>q: Si estudio aprobare el ciclo escolar.

* si p entonce q
* si p, q

* p entonces q.
* p es condicion suficiente para q
* q es condicion necesaria para p

### bicondicional o equivalencia

* el simbolo que lo representa es <=>

* si p y q representa dos proposiciones simples entonces la proposicion compuesta:
                                        p <=> q
* **solo sera verdadera cuando ambas proposiciones tengan el mismo valor de verdad**

* algunas formas de condicional son
    si y solo si, entonces y solo entonces, es identico, equivale a, es equivalente a, entre otras mas.


#### ejemplos

* p: si un poligono tiene cuatro lados, entonces es un cuadrilatero
* q: si un poligono es un cuadrilatero entonces tiene cuatro lados
* p <=> q: un poligoono es cuadrilatero si y solo si tiene cuatro lados
